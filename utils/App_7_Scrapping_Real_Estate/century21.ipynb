```python
l = []
for item in all:
    d={}
    
    d["Adress"] =item.find_all("span", {"class", "propAddressCollapse"})[0].text
    d["Locality"] =item.find_all("span", {"class", "propAddressCollapse"})[1].text
    d["Price"] = item.find("h4", {"class":"propPrice"}).text.replace("\n", "").replace(" ","")
    try:
        d["Beds"] = item.find("span", {"class", "infoBed"}).text #divider <span class="infoBed"><b>3</b> Beds</span>
    except:
        d["Beds"] = None
    
    try:
        d["Area"] = item.find("span", {"class", "infoSqFt"}).text #divider <span class="infoBed"><b>3</b> Beds</span>
    except:
        d["Area"] = None
    
    try:
        d["Full Bath"] = item.find("span", {"class", "infoValueFullBath"}).text#divider <span class="infoBed"><b>3</b> Beds</span>
    except:
        d["Full Bath"] = None

    try:
        d["Half Bath"] = item.find("span", {"class", "infoValueHalfBath"}).text #divider <span class="infoBed"><b>3</b> Beds</span>
    except:
        d["Half Bath"] = None
        
    for column_group in item.find_all("div", {"class":"columnGroup"}):
        for feature_group, feature_name in zip(column_group.find_all
                                                   ("span", {"class":"featureGroup"}), column_group.find_all("span", {"class":"featureName"})):
           if "Lot Size" in feature_group.text:
               d["Lot Size"] = feature_name
            #print(feature_group.text, feature_name.text)
    l.append(d)
```

```python
import pandas
df=pandas.DataFrame(l)
```
